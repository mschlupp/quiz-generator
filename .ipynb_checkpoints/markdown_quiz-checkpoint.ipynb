```python
from jupytercards import md2json, display_flashcards
import json
```

```python
cards='''

## Concise Card Style
In this style, the front of the card is used as the card name, and the back of the card
is the content that follows until there is a new level 1 heading or the end of the string.

## Verbose Card Style
In the Verbose Card Style, the front is defined separately from the name and may
have multiple lines of content.

Cards can have mathematics written in LaTex:

$$
 \int e^x~dx
$$
---
The back follows a horizontal rule `---` and may also have 

multiple 

lines of content
'''

```

```python


myjson=md2json(cards , "cards.json")
print(myjson)

```

```python
quiz = '''

# This is how you would structure a question, using the pound sign.
type: mc - this is the type
* Each potential answer is started with an asterisk.
-c t
-f This is the feedback. It begins with the -f and the line above is the boolean

* Dictionary
-c f
-f Not quite

* Tuple
-c t
-f While it is similar, a tuple is made with ()

# Notice below that a question's type can also just be the letters
mc
* True
-c f
-f Incorrect

* False
-c t
-f Correct

# Notice below that a question's type can also be the type name in JupyterQuiz
multiple_choice
* True
-c f
-f Incorrect

* False
-c t
-f Correct
'''
```

```python
def md2json(md):
    quiz_types = {"multiple_choice": ["mc", "multiple_choice"],
                  "many_choice": ["mnc", "mmc", "many_choice"],
                 "number": ["number", "n"]
                    }
    x=0
    lines = md.splitlines()
    quiz_data = []
    for line in lines:
        line = line.strip()
        if line:
            if line[0] == "#":
                question_data = {}
                
                question = line[1:].strip()
                question_data["question"] = question
                #find type of quiz
                q = lines[x+1]
                
                for quiz in quiz_types:
                    for item in quiz_types[quiz]:
                        if item in q:
                            type_quiz = quiz
                answers = "\n".join(lines[x+2:]).split("#")[0]
                answers = answers.split("*")[1:]
                answer_data = []
                corrects= 0
                for answer in answers:
                    answer = answer.strip()
                    answer, correct, feedback = answer.splitlines()[:3]
                    if "f" in correct.lower():
                        correct = False
                    elif "t" in correct.lower():
                        correct = True
                    if "-f" == feedback[:2]:
                        feedback = feedback[2:].strip()
                    if correct == True:
                        corrects = corrects+1
                    answer_data.append(
                        {"answer": answer,
                        "correct": correct,
                         "feedback": feedback
                         })
                    if corrects == 1:
                        type_quiz = "multiple_choice"
                    elif corrects > 1:
                        type_quiz = "many_choice"
                quiz_data.append({"question": question,
                                 "type": type_quiz,
                               "answers": answer_data})
                
        x=x+1
    
    return json.dumps(quiz_data, indent=4)
myjson_quiz = md2json(quiz)
print (myjson_quiz)
```
